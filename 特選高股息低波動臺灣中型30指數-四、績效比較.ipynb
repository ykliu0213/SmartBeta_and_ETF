```python
ETF_net_worth_list = []
Active_date_net_worth = []
total_buy_amount = []
total_buy_cost = []
total_sell_amount = []
total_sell_cost = []
net_worth = 900000000
periods = 0
for i in range(len(Date_list)):
    if i == 0 :
        # 第一期
        # 計算成分股權重
        temp = df_Dmergefinal_ETF[ df_Dmergefinal_ETF['年月日']==Date_list[i] ]
        # 30檔成分股資料 : temp1
        temp1 = temp[(temp['New_Remain_Del']==0)|(temp['New_Remain_Del']==1)]
        
        # 計算各股持有股數
        shares_list = []
        for j in range(len(temp1)):
            Number_of_shares = math.ceil(( temp1['Weight'].iloc[j] * net_worth ) / temp1['收盤價'].iloc[j] )
            if(Number_of_shares % 1000) != 0 : 
                shares_list.append((int(Number_of_shares/1000) + 1) * 1000)
        # 將各股持有股數合併到temp1
        temp1['Shares'] = shares_list
        
        # 用來暫時紀錄本期的成分股股數，以便生效日出售用
        shares_temp = temp1[['證券代碼','Shares']]
        
        
        total_buy_each = 0
        # 總購買金額 （股數 * 收盤價）
        total_buy_each = sum( temp1['Shares'] * temp1['收盤價'] )
        # 將總賣出金額與賣出成本分別加到list中
        buy_cost = (total_buy_each * 0.001425 * 0.3)
        total_buy_amount.append(total_buy_each)
        total_buy_cost.append(buy_cost)
        
        # 將淨值扣除成本
        net_worth = net_worth - buy_cost
        
        # 記錄每日淨值
        ETF_net_worth_list.append(net_worth)
        
        # 紀錄生效日淨值
        Active_date_net_worth.append(net_worth)
        
        print(net_worth)
        
    elif (Date_list[i] in Active_Date_list):
        # 指數生效日（6/19 or 12/19）
        # 調整成分股權重
        temp            = df_Dmergefinal_ETF[ df_Dmergefinal_ETF['年月日']==Date_list[i]]
        # 30檔成分股資料 : temp1
        temp1           = temp[(temp['New_Remain_Del']==0)|(temp['New_Remain_Del']==1)]

        # 賣出上期持有，但本期不繼續持有之成分股
        total_sell_each = 0
        for j in range(len(Sell_list[periods])):
            # 股數
            shares_each      = (Sell_list[periods][j],shares_temp[shares_temp['證券代碼'] == Sell_list[periods][j]]['Shares'].iloc[0])
            # 收盤價
            close_price_each = (df_Dmergefinal_ETF[(df_Dmergefinal_ETF['證券代碼']==Sell_list[periods][j]) & (df_Dmergefinal_ETF['年月日']==Date_list[i])]['收盤價'].iloc[0])
            # 總出售金額
            total_sell_each += ( shares_each[1] * close_price_each )
        # 將總賣出金額與賣出成本分別加到list中
        sell_cost = ((total_sell_each * 0.001425 * 0.3) + (total_sell_each * 0.003))
        total_sell_amount.append(total_sell_each)
        total_sell_cost.append(sell_cost)
        
        
        # 重新計算各股持有股數
        shares_list = []
        for j in range(len(temp1)):
            Number_of_shares = math.ceil(( temp1['Weight'].iloc[j] * net_worth ) / temp1['收盤價'].iloc[j] )
            if(Number_of_shares % 1000) != 0 : 
                shares_list.append((int(Number_of_shares/1000) + 1) * 1000)
        # 將各股持有股數合併到temp1
        temp1['Shares'] = shares_list
        
        # 用來暫時紀錄本期的成分股股數，以便生效日出售用
        shares_temp = temp1[['證券代碼','Shares']]
        
        
        # 新增的成分股，計算期購入成本
        # temp1['New_Remain_Del']==0
        # 總購買金額 （股數 * 收盤價）
        total_buy_each = sum( temp1[temp1['New_Remain_Del']==0]['Shares'] * temp1[temp1['New_Remain_Del']==0]['收盤價'] )
        # 將總購買金額與購買成本分別加到list中
        buy_cost = (total_buy_each * 0.001425 * 0.3)
        total_buy_amount.append(total_buy_each)
        total_buy_cost.append(buy_cost)
        
        
        #print('total_buy_each : ' , total_buy_each , '    total_sell_each : ' , total_sell_each)
        #print('adj : ', (total_sell_each - total_buy_each))
        
        
        # 計算每日 net_worth （股數 * 收盤價）
        net_worth = sum( temp1['Shares'] * temp1['收盤價'] )
        # 調整淨值，加上賣出金額再減掉買入金額
        net_worth = net_worth + (total_sell_each - total_buy_each)
        # 將淨值扣除成本
        net_worth = net_worth - buy_cost - sell_cost
        
        # 查看淨值
        print(i,net_worth)
        
        # 記錄每日淨值
        ETF_net_worth_list.append(net_worth)
        
        # 紀錄生效日淨值
        Active_date_net_worth.append(net_worth)
        
        # 處理完當期調整，期數+1
        periods = periods + 1

        
        
    else:
        # 其它天
        temp            = df_Dmergefinal_ETF[ df_Dmergefinal_ETF['年月日']==Date_list[i] ]
        # 30檔成分股資料 : temp1
        temp1           = temp[(temp['New_Remain_Del']==0)|(temp['New_Remain_Del']==1)]
        # 合併原始資料和各股股數
        temp1 = pd.merge(left=temp1,right=shares_temp,left_on=['證券代碼'],right_on=['證券代碼'],how='left')
        
        
        
        # 計算每日 net_worth （股數 * 收盤價）
        net_worth = sum( temp1['Shares'] * temp1['收盤價'] )
        
        # 記錄每日淨值
        ETF_net_worth_list.append(net_worth)
```
